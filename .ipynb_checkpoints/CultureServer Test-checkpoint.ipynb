```python
# %load CultureServer.py
import rpyc 
import datetime, time, threading
import fileinput

from ctypes import *
from dwfconstants import *
import math
import time
import matplotlib.pyplot as plt
import sys

import fitSine as fs
import numpy as np

import dwf

#dwInitialzie import ok then AD2 is connected
AD2isConnected = True

MAX_MESSAGE_LINES = 100
timefmt = '%Y-%m-%d %H:%M:%S'


class CultureService(rpyc.Service):
  finishRequest = False
  Status = {
            'isRecording': False,
            'isAD2Connected': False,
            'isBusy': False
    }
  RecordingInfo = {
            'isPaused': False,
            'channels': [],
            'freqs': [],
            'period': 0,
            'deadline': 0
    }
  recordingFname = ""
  noC = 0
  noF = 0
  dline = datetime.datetime.now()
  channels = []
  freqs = []
  period = 0
  deadline=0

  if AD2isConnected == True: Status['isAD2Connected']=True #if CultureService is starting, then AD2 is connected

  def saveMessage(self,msg):
    f=open('msg.txt','a')
    cur=datetime.datetime.now()
    f.write(cur.strftime(timefmt+'---')+msg+'\n')
    f.close()
    num_lines = sum(1 for line in open('msg.txt'))
    if num_lines > MAX_MESSAGE_LINES:
        for line in fileinput.input('msg.txt',inplace=True):
            if fileinput.lineno() == 1:
                continue
            print line[0:-1]   #delete first line only
    fileinput.close()
  
  def on_connect(self):
    print 'server connected'
    self.saveMessage("Sever is connected!")

  def on_disconnect(self):
    print 'server disconnected'
    self.saveMessage("Sever is disconnected!")

  def exposed_finish(self):
    if self.Status['isRecording'] == True:
        print 'finish request arrived'
        self.finishRequest = True
    else:
        self.saveMessage('finish request--Culture Server is not on recording!')
        print 'finish request--Culture Server is not on recording!'

  def exposed_pause(self):
    #recording and measure pause
    if self.Status['isRecording'] == True:
        print 'pause request arrived'
        self.RecordingInfo['isPaused']=True
    else:
        self.saveMessage('pause request--Culture Server is not on recording!')
        print 'pause request--Culture Server is not on recording!'

  def exposed_getStatus(self): # this is an exposed method 
    return self.Status 

  def exposed_getRecordingInfo(self): # this is an exposed method 
    return self.RecordingInfo

  def exposed_setup(self,info): # this is an exposed method 
    if self.Status['isRecording'] == False:
        print 'setup request arrived'
        self.RecordingInfo['isPaused'] = info['isPaused']
        self.RecordingInfo['channels'] = info['channels']
        self.RecordingInfo['freqs'] = info['freqs']
        self.RecordingInfo['period'] = info['period']
        self.RecordingInfo['deadline'] = info['deadline']
    else:
        self.saveMessage('Culture Server is on recording! Finish first')
        print 'Culture Server is on recording! Finish first'

  def checkRecordingInfo(self):
    if (self.RecordingInfo['isPaused']==True): return False
    if (self.RecordingInfo['channels']==[]): return False
    if (self.RecordingInfo['freqs']==[]): return False
    if (self.RecordingInfo['period']==0): return False
    if (self.RecordingInfo['deadline']==0): return False
    return True
  
  def exposed_checkChip(self):
    if self.Status['isRecording'] == False:
        print 'check chip request arrived'
        channels = range(8) # all channels
        freqs=[4000] # default frequency
        Z=dwf.measureImpedance(channels,freqs)
        return Z
    else:
        self.saveMessage('checkChip--Culture Server is on recording! Finish first')
        print 'checkChip--Culture Server is on recording! Finish first'

  def exposed_getScopeData(self, ch, fr):
    if self.Status['isRecording'] == False:
        print 'scope data request arrived'
        self.Status['isBusy']=True
        data0, data1 = dwf.getScopeData(ch, fr)
        self.Status['isBusy']=False
        return data0, data1
    else:
        self.saveMessage('get scope data--Culture Server is on recording! Finish first')
        print 'get scope data--Culture Server is on recording! Finish first'

  def lineout(self,f, comment, arr):
      outs=comment+','
      for i in range(len(arr)):
        outs += str(arr[i])+','
      print>>f, outs

  def exposed_recording(self):
    if self.Status['isRecording'] == False:
        if (self.checkRecordingInfo() == False):
            self.Status['isRecording'] = False
            return
        else: 
            #now start measure and recording
            self.Status['isRecording'] = True
            cur = datetime.datetime.now()
            self.recordingFname = cur.strftime('%Y%m%d_%H%M%S')+'.txt'
            # dictionary should be expanded because the dictionary seems unsafe
            self.dline = cur+datetime.timedelta(minutes=self.RecordingInfo['deadline']) # Release--when debugging, deadline = minutes, change to days
            self.channels = list(self.RecordingInfo['channels'])
            self.freqs = list(self.RecordingInfo['freqs'])
            self.period = self.RecordingInfo['period']
            self.noC=len(self.RecordingInfo['channels'])
            self.noF=len(self.RecordingInfo['freqs'])
            f=open(self.recordingFname,'w')
            self.lineout(f,'channels', self.RecordingInfo['channels'])
            self.lineout(f,'freqs', self.RecordingInfo['freqs'])
            self.lineout(f,'period', [self.RecordingInfo['period']]) #schalar
            self.lineout(f,'deadline', [self.RecordingInfo['deadline']]) #schalar
            f.close()
            self.saveMessage("Recording is started")
            
    now = datetime.datetime.now()
    #Release correction required
    if (now < self.dline) and ( self.finishRequest == False):
       if self.RecordingInfo['isPaused'] == False:
          #measure(channels, Freqs) and recording
          self.Status['isBusy']=True
          Z=dwf.measureImpedance(self.channels,self.freqs)
          #Z = [[(35207.834113369237-41605018j)], [(35188.752862288602-80762649j)], [(35194.685116298249-36166343j)]]
          f=open(self.recordingFname,'a')
          cur = datetime.datetime.now()
          print>>f, cur.strftime(timefmt)
          for ic in range(self.noC): #range(len(self.RecordingInfo['channels'])):
            self.lineout(f,'',Z[ic])
          f.close()
          self.Status['isBusy']=False
          print cur
          t = threading.Timer(self.period,self.exposed_recording)
          t.start()
    else:
       if (now >= self.dline): self.saveMessage('deadline passed and finished!')
       if ( self.finishRequest == True): 
           self.saveMessage('finish request arrive finished!')
           self.finishRequest = False
       self.Status['isRecording']=False
       self.RecordingInfo['isPaused']=False

if __name__ == "__main__": 
  from rpyc.utils.server import ThreadedServer 
  t = ThreadedServer(CultureService, port = 18861) 
  t.start()
```
