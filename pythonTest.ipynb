```python
import numpy as np
```

```python
y=[]
a=range(1024)
for i in a:
    y.append(np.sin(a[i]*np.pi*2/1024))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a,y)
```

```python
plt.show()
```

```python
for i in a:
    y.append(np.exp(a[i]*np.pi*2/1024*1j))
```

```python
z=[]
for i in a:
    z.append(y[i]*np.conj(y[i]))
    
```

```python
plt.plot(a,z)
```

```python
plt.show()
```
